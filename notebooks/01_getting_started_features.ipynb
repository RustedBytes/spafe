# Getting Started: Cepstral Features

This notebook builds a synthetic tone, extracts several cepstral feature families through the `spafe` Python extension, and visualizes the result with matplotlib.

````python
import math

import matplotlib.pyplot as plt
import spafe
````

````python
fs = 16_000
seconds = 1.0
signal = [math.sin(2.0 * math.pi * 440.0 * n / fs) for n in range(int(fs * seconds))]

opts = spafe.FeatureOptions(fs=fs, nfft=256, nfilts=24, win_hop=0.02)
print(f"samples={len(signal)}, fs={fs}")
````

````python
mfcc = spafe.mfcc(signal, opts)
gfcc = spafe.gfcc(signal, opts)
mel_spec = spafe.mel_spectrogram(signal, opts)

print(f"MFCC shape: {len(mfcc)} x {len(mfcc[0])}")
print(f"GFCC shape: {len(gfcc)} x {len(gfcc[0])}")
print(f"Mel spectrogram shape: {len(mel_spec.features)} x {len(mel_spec.features[0])}")
````

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), constrained_layout=True)
for ax, matrix, title, xlabel in [
    (axes[0], mfcc, "MFCC", "Coefficient"),
    (axes[1], gfcc, "GFCC", "Coefficient"),
    (axes[2], mel_spec.features, "Mel Spectrogram", "Filter"),
]:
    image = ax.imshow(matrix, aspect="auto", origin="lower", cmap="viridis")
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Frame")
    fig.colorbar(image, ax=ax, shrink=0.8)
plt.show()
````

````python
features = spafe.extract_feats(signal, fs, nfft=512)
print(f"spectral centroid: {features.spectral_centroid:.3f}")
print(f"spectral entropy: {features.spectral_entropy:.3f}")
print(f"spectral flatness: {features.spectral_flatness:.6f}")
````
