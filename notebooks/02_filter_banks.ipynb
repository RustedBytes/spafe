# Filter Banks

This notebook compares Mel, linear, Bark, and gammatone filter banks from the Python extension.

````python
import matplotlib.pyplot as plt
import spafe
````

````python
opts = spafe.FilterBankOptions(nfilts=24, nfft=512, fs=16_000, high_freq=8_000.0)
freqs = [idx * opts.fs / opts.nfft for idx in range(opts.nfft // 2 + 1)]

banks = {
    "Mel": spafe.mel_filter_banks(opts)[0],
    "Linear": spafe.linear_filter_banks(opts)[0],
    "Bark": spafe.bark_filter_banks(opts)[0],
    "Gammatone": spafe.gammatone_filter_banks(opts)[0],
}
{key: (len(value), len(value[0])) for key, value in banks.items()}
````

````python
fig, axes = plt.subplots(2, 2, figsize=(14, 8), constrained_layout=True)
for ax, (name, bank) in zip(axes.ravel(), banks.items()):
    for row in bank:
        ax.plot(freqs, row, linewidth=1.0)
    ax.set_title(name)
    ax.set_xlabel("Frequency (Hz)")
    ax.set_ylabel("Weight")
    ax.grid(True, alpha=0.25)
plt.show()
````

````python
_, centers = spafe.mel_filter_banks(opts)
print("first Mel centers:", [round(value, 3) for value in centers[:5]])
print("1000 Hz in Mel:", round(spafe.hz2mel(1000.0), 3))
print("roundtrip:", round(spafe.mel2hz(spafe.hz2mel(1000.0)), 3))
````
